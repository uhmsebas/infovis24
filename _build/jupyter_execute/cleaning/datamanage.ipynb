```python
import pandas as pd
import numpy as np
```

```python
#read in the data
pd_gdp = pd.read_csv('gdp_data.csv')
display(pd_gdp.head())
pd_gdp = pd_gdp.drop(columns='Country Code',axis=1)
```

```python
pd_gdp_pivot = pd_gdp.pivot_table(index='Country Name', columns='Series Name', values='2022 [YR2022]', aggfunc='first')
display(pd_gdp_pivot.head())
```

```python
pd_gdp_pivot.to_csv('gdp_data_pivot.csv')
```

```python
#read data
df_edu = pd.read_csv('API_SE.SEC.CMPT.LO.ZS_DS2_en_csv_v2_655656.csv')
display(df_edu.head())
```

```python
#only keep the 2022 data
df_edu = df_edu[['Country Name','2022']]
display(df_edu)
```

```python
df_merged = pd.merge(df_edu, pd_gdp_pivot, on='Country Name', how='inner')

#rename 2022
df_merged.rename(columns={'2022': 'Enrollment 2022'}, inplace=True)
display(df_merged)
```

```python
df_merged.to_csv('merged_with_NAN_edu-gdp-life.csv')
df_merged = df_merged.dropna()
df_merged.to_csv('merged_with_droppedNAN_edu-gdp-life.csv')
```
