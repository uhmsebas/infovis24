```python
import pandas as pd
import numpy as np
```

```python
# Load the data
pd_crime = pd.read_csv('croime.csv')
display(pd_crime)
```

```python
# find columns with the least missing values

# Count NaN values in each year column
pd_drop = pd_crime.drop(columns=['Country Name','Country Code','Series Name','Series Code'])
# change .. to NaN  
pd_drop = pd_drop.replace('..', np.nan)

df_list = pd_drop.isna().sum()

# Find the minimum NaN value and the corresponding year
min_nan_count = df_list.min()
year_with_min_nan = df_list.idxmin()
#print the result

print(f"Year with the least missing values: {year_with_min_nan} with {min_nan_count} missing values")
```

```python
pd_crime = pd_crime[['Country Name','Series Name',year_with_min_nan]]
display(pd_crime)
```

```python
#pivot the series name with year as dat
pd_crime = pd_crime.pivot(index=['Country Name'], columns='Series Name', values=year_with_min_nan)
display(pd_crime)
```

```python
#writing the result to a csv file
pd_crime.to_csv('crime_pivot.csv')
```

```python
pd_child = pd.read_csv('chil.csv')
pd_child.rename(columns={'2010 [YR2010]':'Mortality rate, infant (per 1,000 live births)'}, inplace=True)
pd_child = pd_child[['Country Name','Mortality rate, infant (per 1,000 live births)']]
display(pd_child)
```

```python
#merge the two dataframes
pd_merge = pd.merge(pd_crime, pd_child, on='Country Name', how='inner')
display(pd_merge)
```

```python
pd_merge.to_csv('crime_child_merge.csv')
```

```python
pd_disease = pd.read_csv('disease.csv')
pd_disease.rename(columns={'2010 [YR2010]':'communicable diseases'}, inplace=True)
pd_disease = pd_disease[['Country Name','communicable diseases']]
display(pd_disease)
```

```python
#merge the two dataframes
pd_merge_disease = pd.merge(pd_merge, pd_disease, on='Country Name', how='inner')
display(pd_merge_disease)
```

```python
pd_merge_disease.to_csv('crime_child_disease_merge.csv')
```

```python
df = pd.read_csv('all_vaccine_data.csv')
df = df.drop(columns=['DOSES', 'TARGET_NUMBER'])
df = df.dropna()
df = df[['YEAR','ANTIGEN_DESCRIPTION','COVERAGE','Country Name','continent','Life Expectancy','GDP per Capita','Child Mortality per 1000 ']]
#pivot the antigen_description column to get the antigens as columns with the coverage as the values
df = df.pivot_table(index=['YEAR','Country Name','continent','Life Expectancy','GDP per Capita','Child Mortality per 1000 '], columns='ANTIGEN_DESCRIPTION', values='COVERAGE')
df = df.reset_index()
display(df)
df.to_csv('vaccine_pivot.csv')
```
