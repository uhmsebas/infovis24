```python
import pandas as pd
import numpy as np
import pycountry
```

```python
# Read the data from the CSV and excel file
df_polio = pd.read_excel(r'polio.xlsx', engine='openpyxl')
df_measles = pd.read_excel('mesles.xlsx', engine='openpyxl')
df_diphtheria =pd.read_excel('Diphtheria.xlsx', engine='openpyxl')
df_contnent = pd.read_csv('list-of-countries-by-continent-2024.csv')
display(df_polio)
display(df_measles)
display(df_diphtheria)
display(df_contnent)

```

```python
#concatenate the dataframes
df_illness = pd.concat([df_polio, df_measles, df_diphtheria])
display(df_illness)
df_illness.to_csv('allillnesses.csv', index=False)
```

```python
#change the country names to the standard names
def standardize_country_name(country_name):
    try:
        # Attempt to get the standard country name
        country = pycountry.countries.lookup(country_name)
        return country.name
    except LookupError:
        # Return the original name if no match is found
        return country_name

# Apply the function to the datasets
df_illness['standardized_country'] = df_illness['NAME'].apply(standardize_country_name)
df_contnent['standardized_country'] = df_contnent['country'].apply(standardize_country_name)
#drop replaced columns
df_illness = df_illness.drop(columns=['NAME'])
df_contnent = df_contnent.drop(columns=['country'])
display(df_illness)
display(df_contnent)

```

```python
#merge the dataframes
df_mergde = pd.merge(df_illness, df_contnent, on='standardized_country', how='inner')
display(df_mergde)
#change year to int
df_mergde['YEAR'] = df_mergde['YEAR'].astype(int)
display(df_mergde)
df_mergde.to_csv('vaccinated_continets.csv', index=False)
```

```python
pd_life = pd.read_csv('life_expactancy_2_worldbank.csv')  
pd_life.drop(columns=['Unnamed: 68'], inplace=True) 
display(pd_life)
```

```python
#meltinf the data
pd_life_melted = pd.melt(pd_life, id_vars=['Country Name', 'Country Code','Indicator Name','Indicator Code'], var_name='Year', value_name='Life Expectancy')
display(pd_life_melted)
```

```python
#normalize the country names
pd_life_melted['standardized_country'] = pd_life_melted['Country Name'].apply(standardize_country_name)
pd_life_melted = pd_life_melted.drop(columns=['Country Name'])	
display(pd_life_melted)
```

```python
# make year to int
pd_life_melted['Year'] = pd_life_melted['Year'].astype(int)
#merge the dataframes
df_merged_life = pd.merge(df_mergde, pd_life_melted, left_on=['standardized_country','YEAR'],right_on=['standardized_country','Year'], how='inner')
#rename the columns
df_merged_life.rename(columns={'standardized_country':'Country Name'}, inplace=True)
display(df_merged_life)
```

```python
#drop irrelevant columns
df_merged_life.to_csv('undropped_better_merge.csv', index=False)
df_merged_life.drop(columns=['Country Code','Indicator Name','Indicator Code','COVERAGE_CATEGORY', 'COVERAGE_CATEGORY_DESCRIPTION', 'GROUP', 'Year'], inplace=True)
display(df_merged_life)
```

```python
#write the data to a csv file
df_merged_life.to_csv('all_vaccane_1980-2023.csv', index=False)
```

# extra merging for gdp and infant death


```python
pd_gdp = pd.read_csv('gdpcapita.csv')
#drop unnamed column   
pd_gdp.drop(columns=['Unnamed: 68'], inplace=True)
display(pd_gdp)
pd_child = pd.read_csv('Kinderdoden.csv')
#drop unnamed column
pd_child.drop(columns=['Unnamed: 68'], inplace=True)
display(pd_child)


```

```python
#meltinf the data
pd_gdp_melted = pd.melt(pd_gdp, id_vars=['Country Name', 'Country Code','Indicator Name','Indicator Code'], var_name='Year', value_name='GDP')
display(pd_gdp_melted)
pd_gdp__meltd_drop = pd_gdp_melted.drop(columns=['Country Code','Indicator Name','Indicator Code'])
#equalize the country names
pd_gdp__meltd_drop['standardized_country'] = pd_gdp__meltd_drop['Country Name'].apply(standardize_country_name)
pd_gdp__meltd_drop = pd_gdp__meltd_drop.drop(columns=['Country Name'])
#rename the columns
pd_gdp__meltd_drop.rename(columns={'standardized_country':'Country Name'}, inplace=True)
#make year to int
pd_gdp__meltd_drop['Year'] = pd_gdp__meltd_drop['Year'].astype(int)
display(pd_gdp__meltd_drop)
```

```python
#the same for the child mortality
pd_child_melted = pd.melt(pd_child, id_vars=['Country Name', 'Country Code','Indicator Name','Indicator Code'], var_name='Year', value_name='Child Mortality')
display(pd_child_melted)
pd_child_melted_drop = pd_child_melted.drop(columns=['Country Code','Indicator Name','Indicator Code'])
pd_child_melted_drop['standardized_country'] = pd_child_melted_drop['Country Name'].apply(standardize_country_name)
pd_child_melted_drop = pd_child_melted_drop.drop(columns=['Country Name'])
pd_child_melted_drop.rename(columns={'standardized_country':'Country Name'}, inplace=True)
#rename the columns
pd_child_melted_drop.rename(columns={'Child Mortality':'Child Mortality per 1000 '}, inplace=True)
pd_child_melted_drop['Year'] = pd_child_melted_drop['Year'].astype(int)
display(pd_child_melted_drop)
```

```python
#merge the dataframes 
df_merged_gdp = pd.merge(df_merged_life, pd_gdp__meltd_drop, left_on=['Country Name','YEAR'],right_on=['Country Name','Year'], how='inner')
#drop the year column
df_merged_gdp.drop(columns=['Year'], inplace=True)
display(df_merged_gdp)
df_merged_gdp_child = pd.merge(df_merged_gdp, pd_child_melted_drop, left_on=['Country Name','YEAR'],right_on=['Country Name','Year'], how='inner')
#drop the year column
df_merged_gdp_child.drop(columns=['Year'], inplace=True)
#renaem the gdp column
df_merged_gdp_child.rename(columns={'GDP':'GDP per Capita'}, inplace=True)
display(df_merged_gdp_child)
df_merged_gdp_child.to_csv('all_vaccine_data.csv', index=False)
```

```python
#put each diffferent illness in a different colu
list_of_indicators = df_merged_gdp_child['ANTIGEN_DESCRIPTION'].unique()
for indicator in list_of_indicators:
    df_merged_gdp_child[indicator] = np.where(df_merged_gdp_child['ANTIGEN_DESCRIPTION'] == indicator, df_merged_gdp_child['COVERAGE'], np.nan)
df_merged_gdp_child.drop(columns=['ANTIGEN_DESCRIPTION', 'COVERAGE','ANTIGEN'], inplace=True)
display(df_merged_gdp_child)
df_merged_gdp_child.to_csv('errorvaccinne.csv', index=False)
```

```python
#join rows of the same country in the same year
df_merged_gdp_child_grouped = df_merged_gdp_child.groupby(['Country Name', 'YEAR']).max().reset_index()
display(df_merged_gdp_child_grouped)
df_merged_gdp_child_grouped.to_csv('final_vaccine_data_columns.csv', index=False)
```
